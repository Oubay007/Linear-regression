```python
import pandas as pd

from google.colab import files
import zipfile
import io
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, mean_squared_error, confusion_matrix, classification_report



uploaded = files.upload()

zip_filename = list(uploaded.keys())[0]

# Unzip the uploaded file in memory and extract all
with zipfile.ZipFile(io.BytesIO(uploaded[zip_filename]), 'r') as zip_ref:
    zip_ref.extractall()  # Extracts to current directory

df = pd.read_csv('spam.csv')

def clean_txt(text) :
  import re, string
  text = text.lower()
  text = re.sub('<.*?>' , ' ', text)
  text = re.sub('[%s]' % re.escape(string.punctuation) , ' ', text)
  return text

df['text'] = df['Message'].apply(clean_txt)
df.head(20)
df.drop(columns = ['Message'])

vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['text'])

df['Label'] = df['Category'].map({'spam' : 1, 'ham' : 0})
y = df['Label']

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.2)

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_val)

accuracy = accuracy_score(y_val, y_pred)
mse = mean_squared_error(y_val, y_pred)

# Display results
print("âœ… Accuracy:", round(accuracy * 100, 2), "%")
print("ðŸ“‰ Mean Squared Error (MSE):", round(mse, 4))

#test :
new_email = """
Hi John,

I hope youâ€™re doing well. Just wanted to check if youâ€™re available for the meeting tomorrow at 3 PM.

Best regards,
Sarah
"""

cleaned_email = clean_txt(new_email)


email_vector = vectorizer.transform([cleaned_email])


prediction = model.predict(email_vector)


label_map = {0: "ham", 1: "spam"}
print(f"Prediction: {label_map[prediction[0]]}")

```
